```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Читаем CSV
url = "https://raw.githubusercontent.com/lancer43/SOR-1/main/vs-project/AnalysisSortAlgorithms/results.csv"
df = pd.read_csv(url, delimiter=';')

# посмотрим на данные
print(df.head())
print("\nСтатистика:")
print(df.describe())

# график
plt.figure(figsize=(14, 8))
for col in df.columns[1:]:
    plt.plot(df['Size'], df[col], label=col, linewidth=1.5, marker='.', markersize=3)

plt.xlabel('Размер массива', fontsize=12)
plt.ylabel('Время (мс)', fontsize=12)
plt.title('Сравнение всех алгоритмов сортировки', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
